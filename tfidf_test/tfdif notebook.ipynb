```python
def remove_string_special_char(s):
    
    """This function removes special chars from within a string"""
    
    # Replace special chars with ''
    stripped = re.sub('[^\w\s]', '', s)
    stripped = re.sub('_', '', stripped)
    
    #change any whitespace to one space
    stripped = re.sub('\s+', '', stripped)
    
    #remove start and end white spaces
    stripped = stripped.strip()
    
    return stripped
```

```python
    def create_freq_dict(sents):
                
                """This function creates a frequency dictionary
                for each word in each document"""
                
                i = 0
                freq_dict_list = []
                for sent in sents:
                    i += 1
                    freq_dict = {}
                    words = word_tokenize(sent)
                    for word in words:
                        word = word.lower()
                        if word in freq_dict:
                            freq_dict[word] += 1
                        else: 
                            freq_dict[word] = 1
                        temp = {'doc_id': i, 'freq_dict' : freq_dict}
                    freq_dict_list.append(temp)
                return freq_dict_list
```

```python
def compute_TF(doc_info, freq_dict_list):
    """" tf = number of times the term appears in the doc/total number of words in the doc"""
    TF_scores = []
    for tempDict in freq_dict_list:
        id = tempDict['doc_id']
        for k in tempDict['freq_dict']:
            temp = {'doc_id': id,
                'tf_score' : tempDict['freq_dict'][k]/doc_info[id - 1]['doc_length'],
                    'key': k}
            TF_scores.append(temp)
        return TF_scores
```

```python
def computeTFDIF(TF_scores, IDF_scores):
    TFDIF_scores = []
    for idf in IDF_scores:
        for tf in TF_scores:
            if idf['key'] == tf['key'] and idf['doc_id'] == tf['doc_id']:
                temp = {'doc_id' : idf['doc_id'],
                       'TFDIF_scores' : tf['tf_score']*idf['idf_scores'],
                       'key': tf['key']}
                TFDIF_scores.append(temp)
                return TFDIF_scores
```
